## 1. Data Cleaning

### Trips Table

```python
pd.set_option('display.max_rows', None)

trips.head()
```

```python
## mixed data type check
for col in trips.columns.tolist():
  weird = (trips[[col]].applymap(type) != trips[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (trips[weird]) > 0:
    print (col)
```

```python
## missing observations?

trips.isnull().sum()
```

```python
## dropping customer plan column as it is empty
trips.drop('customer_plan', axis=1, inplace=True)
```

```python
## duplicates?

trips[trips.duplicated()]
```

### Stations Table 

```python
stations.head()
```
